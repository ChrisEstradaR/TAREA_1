```
print("        AREA Y VOLUMEN  ")

def area (rad):
  ar = (rad * rad) * 3.1416
  print(" AREA DEL CIRCULO : ", ar)

def volumen (altura, rad):
  alt = (rad**2) * 3.1416  * altura
  print(" VOLUMEN DEL CILINDRO : ", alt)

a = float (input(" INGRESAR RADIO DEL CIRCULO :"))
b = float (input(" INGRESAR ALTURA DEL CILINDRO :"))
area(a)
volumen(b, a)
```
