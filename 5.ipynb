```
print("         PESO POR PEDIDO DE JUGUETES ")

def peso (payaso, mesa) :
  res1 = payaso * 150
  res2 = mesa * 170
  res3 = res1 + res2 
  print(" PESO DE LOS PAYASOS :", res1,"gr")
  print(" PESO DE LOS JUEGOS DE MESA :", res2,"gr")
  print(" PESO TOTAL DE MEDIDO:", res3,"gr")

  if res3 > 10000 :
    print("       ERROR   EXCESO  DE  PESO")
    res4 = res3 - 10000
    print(" PESO SOBRANTE = ", res4,"gr")
  else :
    print("       PEDIDO  CON  EXITO ")

a = int (input(" INGRESAR LA CANTIDAD DE PAYASOS :"))
b = int (input(" INGRESAR LA CANTIDAD DE JUEGOS DE MESA :"))

peso (a, b)
```
