```
print("    IMPUESTO SALARIAL ")
def impuesto (salario):
  if salario < 10000 :
    print(" IMPUESTO DEDUCIDO DEL 5%")
    imp1 = salario * (5/100)
    print(" SALARIO DESCONTADO :$", imp1 )
    res1 = salario - imp1
    print(" SALARIO TOTAL :$", res1)
  if salario >=10000 and salario < 15000 :
    print(" IMPUESTO DEDUCIDO DEL 10%")
    imp2 = salario * (10/100)
    print(" SALARIO DESCONTADO :$", imp2 )
    res2 = salario - imp2
    print(" SALARIO TOTAL :$", res2)
  if salario >=15000 and salario < 20000 :
    print(" IMPUESTO DEDUCIDO DEL 15%")
    imp3 = salario * (15/100)
    print(" SALARIO DESCONTADO :$", imp3 )
    res3 = salario - imp3
    print(" SALARIO TOTAL :$", res3)
  if salario >=20000 :
    print(" IMPUESTO DEDUCIDO DEL 18%")
    imp4 = salario * (18/100)
    print(" SALARIO DESCONTADO : $", imp4 )
    res4 = salario - imp4
    print(" SALARIO TOTAL : $", res4)
a = float (input(" INGRESAR SALARIO : "))
impuesto(a)
```
